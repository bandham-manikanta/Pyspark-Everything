```python
from pyspark.sql import SparkSession

# Create a Spark session and configure the BigQuery connector
spark = SparkSession.builder.master('local[*]') \
    .appName("BigQueryPublicDatasets") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

# Load data from the Shakespeare dataset
df = spark.read.parquet('noaa_gsod.parquet')

df.show()
```

```python
df.count()
```

```python
df.write.mode('overwrite').format("csv") \
    .option("header", "true").save("noaa_gsod.csv")
```

```python
spark.stop()
```
