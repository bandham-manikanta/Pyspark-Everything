```python
from pyspark.sql import SparkSession
import logging

logging.basicConfig(level=logging.INFO)

# Create a Spark session and configure the BigQuery connector
spark = SparkSession.builder.master('local[*]') \
    .appName("BigQueryPublicDatasets") \
    .config("spark.jars.packages", "com.google.cloud.spark:spark-bigquery-with-dependencies_2.12:0.21.1") \
    .getOrCreate()
```

```python
df = spark.read \
    .format("bigquery") \
    .option("table", "bigquery-public-data.github_repos.commits") \
    .load()

df.createOrReplaceTempView('commits')

df = spark.sql('select * from commits limit 30000')

df.write.mode('overwrite').parquet(r"D:/git_commits.parquet")
```
