```python
file_path = r"Online Retail.csv"

from pyspark.sql import SparkSession

spark = SparkSession\
    .builder.appName("Datacamp Pyspark Tutorial")\
        .getOrCreate()

df = spark.read.csv(file_path,header=True,escape="\"")
df.show(5,0)
```

```python
df.count()
```

```python
df.select('CustomerID').distinct().count()
```

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *

df.groupBy('country').agg(countDistinct('CustomerID').alias('country_count')).show()
```

```python
df.groupBy('Country')\
    .agg(countDistinct('CustomerID').alias('country_count'))\
        .orderBy(desc('country_count')).show()
```

```python
spark.sql("set spark.sql.legacy.timeParserPolicy=LEGACY")
df = df.withColumn('date',to_timestamp("InvoiceDate", 'MM/dd/yyyy HH:mm'))
df.select(max("date")).show()
```

```python
df.select(min("date")).show()
```
